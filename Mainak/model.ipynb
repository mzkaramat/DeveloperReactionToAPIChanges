```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vocabulary=['backward compat','direct api access','free api','good document','good qualiti'
           ,'minimum intervent','nice api','open sourc','rich inform','use inform','authent','use api'
            ,'secur','simplic','simpl','break api chang','break document api','chang term servic'
            ,'charge much money','cross refer paramat','cut off access','function anymor'
            ,'discov new version','inabl cross refer','leak lot implement','limit access api'
            ,'privaci issu','restrict develop access','return format']
```

```python
import pandas as pd
load_data=pd.read_csv("labelled_ycombi.csv")
load_data=load_data[load_data['sentiment']!='neutral']
```

```python
len(load_data)
```

```python
data_post=load_data[['processed_post','sentiment']]
print(data_post['sentiment'].unique())
data_post['sentiment']=data_post['sentiment'].apply(lambda x:1 if x=='postive' else -1)
```

```python
data_list_pos=data_post[data_post['sentiment']==1]['processed_post'].tolist()
```

```python
data_list_neg=data_post[data_post['sentiment']==-1]['processed_post'].tolist()
```

```python
len(data_list_neg)
```

```python
vectorizer1 = CountVectorizer(ngram_range=(1,3),binary=True)
vectorizer1.fit(vocabulary)
```

```python
print(vectorizer1.get_feature_names())
```

```python
vectorizer2 = CountVectorizer(binary=True,vocabulary=vectorizer1.get_feature_names())
term_doc=vectorizer2.transform(data_list_pos)
term_doc=term_doc.toarray()
```

```python
term_doc[0:5]
```

```python
tf_idf_df=pd.DataFrame(data=term_doc,    # values
              columns=vectorizer2.get_feature_names())  # 1st row as the column names
```

```python
tf_idf_df['sentiment']=1
```

```python
term_doc_neg=vectorizer2.transform(data_list_neg)
term_doc_neg=term_doc_neg.toarray()
```

```python
tf_idf_df_neg=pd.DataFrame(data=term_doc_neg,    # values
              columns=vectorizer2.get_feature_names())  # 1st row as the column names
```

```python
tf_idf_df_neg['sentiment']=-1
```

```python
tf_idf_df_neg.head()
```

```python
tf_idf_df_pos_neg=pd.concat([tf_idf_df,tf_idf_df_neg])
tf_idf_df_pos_neg=tf_idf_df_pos_neg.reset_index(drop=True)
```

```python
tf_idf_df_pos_neg = tf_idf_df_pos_neg.sample(frac=1).reset_index(drop=True)
```
