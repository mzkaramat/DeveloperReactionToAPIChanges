```python
import pandas as pd
```

```python
def load_csv_to_df(file_name):
    df=pd.DataFrame.from_csv(file_name)
    df=df.dropna(how="all")
    return df
```

```python
def plot_pivot(df):
    pv = pd.pivot_table(df, index=df.index.year, columns=df["topic"],
                    values='post', aggfunc='count')
    print(pv)
    pv.plot()
```

```python
df1=load_csv_to_df("ycombi_tweet.csv")
df1["topic"]="Twitter"
df2=load_csv_to_df("ycombi_fb.csv")
df2["topic"]="Facebook"
df_total=df1.append(df2)
plot_pivot(df_total)
```
